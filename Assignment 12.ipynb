```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns = ['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
label_encoders = {col: LabelEncoder() for col in categorical_columns}
df_train[categorical_columns] = df_train[categorical_columns].apply(lambda x: label_encoders[x.name].fit_transform(x))
```

```python
df_train
```

```python
df_train.drop(['capitalgain','capitalloss'],axis=1,inplace=True)
```

```python
X_train=df_train.drop(['Salary'],axis=1)
y_train=df_train['Salary']
```

```python
import seaborn as sns
sns.heatmap(X_train.corr())
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
nb_classifier=GaussianNB()
nb_classifier.fit(X_train,y_train)
```

```python
y_pred_train=nb_classifier.predict(X_train)
y_pred_train
```

```python
accuracy=accuracy_score(y_train,y_pred_train)
accuracy
```

```python
print(classification_report(y_train, y_pred_train))
```
