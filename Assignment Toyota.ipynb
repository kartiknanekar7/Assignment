```python
X=df.drop(['Price'],axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=41)
Xtrain.shape
```

```python
Xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_model=LabelEncoder()
le_model.fit(Xtrain['Model'])
Xtrain['Model']=le_model.transform(Xtrain['Model'])

le_Guarantee_Period=LabelEncoder()
le_Guarantee_Period.fit(Xtrain['Guarantee_Period'])
Xtrain['Guarantee_Period']=le_Guarantee_Period.transform(Xtrain['Guarantee_Period'])

le_Color=LabelEncoder()
le_Color.fit(Xtrain['Color'])
Xtrain['Color']=le_Color.transform(Xtrain['Color'])

le_Fuel_Type=LabelEncoder()
le_Fuel_Type.fit(Xtrain['Fuel_Type'])
Xtrain['Fuel_Type']=le_Fuel_Type.transform(Xtrain['Fuel_Type'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
lr=LinearRegression()
lr.fit(Xtrain,ytrain)
```

```python
Xtest['Model']=le_model.fit_transform(Xtest['Model'])
```

```python
Xtest['Guarantee_Period']=le_Guarantee_Period.transform(Xtest['Guarantee_Period'])
```

```python
Xtest['Color']=le_Color.transform(Xtest['Color'])
```

```python
Xtest['Fuel_Type']=le_Fuel_Type.transform(Xtest['Fuel_Type'])
```

```python
y_pred_test=lr.predict(Xtest)
y_pred_test
```

```python
y_pred_train=lr.predict(Xtrain)
y_pred_train
```

```python
from sklearn.metrics import r2_score
score_train=r2_score(y_pred_train,ytrain)
score_test=r2_score(y_pred_test,ytest)
```

```python
score_train,score_test
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
residuals = ytest - y_pred_test
```

```python
sm.qqplot(residuals, line ='r')
plt.title('Q-Q Plot of Residuals')
plt.show()
```

```python
plt.scatter(y_pred_test, residuals)
plt.title('Residuals vs. Predicted Values')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

```python
X_train_const = sm.add_constant(Xtrain)
model_ols = sm.OLS(ytrain, X_train_const).fit()

influence = model_ols.get_influence()
fig, ax = plt.subplots(figsize=(8, 6))
sm.graphics.influence_plot(model_ols, ax=ax, criterion="cooks")
plt.title("Cook's Distance Plot")
plt.show()
```
